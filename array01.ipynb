```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Input 10 nilai mahasiswa
nilai_mahasiswa = [75, 55, 88, 62, 91, 40, 78, 95, 66, 82]

# mencari nilai tertinggi dan terendah
nilai_tertinggi = np.max(nilai_mahasiswa)
nilai_terendah = np.min(nilai_mahasiswa)

# Menghitung rata-rata
rata_rata = np.mean(nilai_mahasiswa)

# menghitung jumlah mahasiswa lulus (>= 60) dan tidak lulus
lulus = len([n for n in nilai_mahasiswa if n >= 60])
tidak_lulus = 10 - lulus
```

```python
print("=== HASIL ANALISIS NILAI MAHASISWA ===")
print(f"Nilai Tertinggi: {nilai_tertinggi}")
print(f"Nilai Terendah: {nilai_terendah}")
print(f"Rata-rata: {rata_rata:.2f}")
print(f"Jumlah Lulus (>= 60): {lulus}")
print(f"Jumlah Tidak Lulus (< 60): {tidak_lulus}\n")
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
kategori = [f'Tertinggi ({nilai_tertinggi})', f'Terendah ({nilai_terendah})']
nilai = [nilai_tertinggi, nilai_terendah]
bars = plt.bar(kategori, nilai, color=['blue', 'red'])
plt.title("VISUALISASI NILAI TERTINGGI & TERENDAH")
plt.ylim(0, 105)

plt.subplot(1, 2, 2)
label_pie = [f'Lulus ({lulus})', f'Tidak Lulus ({tidak_lulus})']
ukuran_pie = [lulus, tidak_lulus]
warna = ['green', 'orange']
plt.pie(ukuran_pie, labels=['Lulus', 'Tidak Lulus'], colors=warna, autopct='%1.1f%%', startangle=90)
plt.title("DATA KELULUSAN MAHASISWA")
plt.legend(label_pie, loc="upper right")

plt.tight_layout()
plt.show()
```
